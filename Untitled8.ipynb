```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Step 1: Generate random data
np.random.seed(42)  # For reproducibility
X = np.random.rand(300, 2)  # 300 data points with 2 features (2D)

# Step 2: Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)  # Assuming 3 clusters
kmeans.fit(X)

# Step 3: Get the cluster centers and labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Step 4: Visualize the results
plt.figure(figsize=(8, 6))

# Plotting the data points with their respective clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.7)

# Plot the cluster centers
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')

plt.title('K-Means Clustering with Random Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

# Step 1: Generate random data
np.random.seed(42)  # For reproducibility
X = np.random.rand(300, 2)  # 300 data points with 2 features (2D)

# Step 2: Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.1, min_samples=5)  # eps is the maximum distance between two samples, min_samples is the minimum number of points to form a cluster
labels = dbscan.fit_predict(X)

# Step 3: Visualize the results
plt.figure(figsize=(8, 6))

# Plotting the data points with their respective clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.7)

# Mark the noise points with a different color
plt.scatter(X[labels == -1, 0], X[labels == -1, 1], c='red', s=100, marker='x', label='Noise')

plt.title('DBSCAN Clustering with Random Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Step 1: Generate random data
np.random.seed(42)  # For reproducibility
X = np.random.rand(300, 2)  # 300 data points with 2 features (2D)

# Step 2: Apply Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=3, linkage='ward')  # Using 3 clusters and 'ward' linkage
labels = agg_clustering.fit_predict(X)

# Step 3: Visualize the clusters
plt.figure(figsize=(8, 6))

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.7)
plt.title('Agglomerative Hierarchical Clustering with Random Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Step 4: Dendrogram to show the hierarchical relationship
linked = linkage(X, 'ward')

plt.figure(figsize=(10, 7))
dendrogram(linked)
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Index')
plt.ylabel('Distance')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Step 1: Generate blobs (clusters of points)
X, y = make_blobs(n_samples=300, centers=3, random_state=42)  # 300 samples, 3 centers

# Step 2: Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Step 3: Visualize the blobs and KMeans results
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.7)

# Plot the cluster centers
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')

plt.title('K-Means Clustering on Blobs')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# Step 1: Generate moons (two interleaving crescent-shaped clusters)
X, y = make_moons(n_samples=300, noise=0.1, random_state=42)  # 300 samples, with some noise

# Step 2: Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.2, min_samples=5)  # You can adjust these parameters
labels = dbscan.fit_predict(X)

# Step 3: Visualize the moons and DBSCAN results
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.7)

# Mark the noise points with a different color (label == -1)
plt.scatter(X[labels == -1, 0], X[labels == -1, 1], c='red', s=100, marker='x', label='Noise')

plt.title('DBSCAN Clustering on Moons')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
